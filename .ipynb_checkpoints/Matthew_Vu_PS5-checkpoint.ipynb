```python
import numpy as np
from math import *
import matplotlib.pyplot as plt

def CalcTaylorSeriesSum(x, N):
    sum = 0
    power = 1
    for i in range(0, N):
        if i % 2 == 1:
            sum -= (x ** power) / (factorial(power))
        else:
            sum += (x ** power) / (factorial(power))
        power += 2
    return sum

def PlotTaylorSeries(x, N):
    x = np.arange(0, x, 0.01)
    y = [CalcTaylorSeriesSum(i, N) for i in x]
    
    ry = np.sin(x)
    
    plt.plot(x, y, 'r-')
    plt.plot(x, ry, 'b-')

N = 85
x = 11*np.pi
PlotTaylorSeries(x, N)
print(CalcTaylorSeriesSum(x, N))
```

```python
def NewtonMethod(func, deriv, iguess, error):
    temp = iguess
    while abs(temp-iguess)<error:
        temp, iguess = iguess, iguess - (func(iguess) / deriv(iguess))
    return iguess
```
