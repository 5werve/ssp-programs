```python
from math import *

def f(i):
    return (sqrt(i**3 + 2*sin(10/i)))

def SolveIntegral(f, a, b, n):
    width = (b-a)/n
    i = a + (width/2)
    result = 0
    
    while i < b:
        result += f(i) * width
        i += width
        
    return result

# About 100 to be within 5 sig figs
# print(SolveIntegral(100))
```

```python
import numpy as np
from math import *

def FindAngle(s, c):
    if (s < 0) and (c < 0):
        return -(np.arcsin(s)) + pi
    elif (s < 0) and (c > 0):
        return (2*pi - (-np.arcsin(s)))
    elif (s > 0) and (c > 0):
        return np.arcsin(s)
    elif (s == 0) and (c == 1):
        return 0
    elif (s == 0) and (c == -1):
        return pi
    elif (c == 0) and (s == 1):
        return pi/2
    elif (c == 0) and (s == -1):
        return 3*pi/2
    else:
        return np.arcsin(s) + (pi/2)
```

```python
from math import *

def DMStoDeg(d, am, asec):
    return ((float(d)) + (copysign(float(am), float(d)) * (1/60)) + (copysign(float(asec), float(d)) * (1/3600)))
DMStoDeg(-60, 50, 0)
```

```python
from math import *

def RAdecimalToHMS(val):
    h = trunc(val / 15)
    val = (abs(val) / 15) - abs(h)
    m = trunc(val / (1/60))
    val = val - (m * (1/60))
    s = val / (1/3600)
    s = round(s, 0)
    if (s == 60):
        s = 0
        m += 1
    print (h, ":", m, ":", s)
    
RAdecimalToHMS(32.5)
```

```python
def DECdecimalToDMS(dec):
    d = trunc(dec)
    dec = abs(dec) - abs(d)
    m = trunc(dec / (1/60))
    dec -= (m * (1/60))
    s = dec / (1/3600)
    s = round(s, 0)
    if (s == 60):
        s = 0
        m += 1
    print (d, ":", m, ":", s)
    
DECdecimalToDMS(-125.32)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

theta = np.arange(0, 10*pi, 0.02)
X = np.cos(theta) * np.sqrt(theta)
Y = np.sin(theta) * np.sqrt(theta)
I = np.cos(theta) * -np.sqrt(theta)
J = np.sin(theta) * -np.sqrt(theta)

plt.plot(X, Y, 'r-')
plt.plot(I, J, 'k-')
plt.show()
```

```python
import numpy as np

def rot3DVector(v, a, b):
    rotMatrix = np.array([[np.cos(a), np.sin(a), 0], [-np.sin(a)*np.cos(b), np.cos(b)*np.cos(a), np.sin(b)], [np.sin(a)*np.sin(b), -np.cos(a)*np.sin(b), np.cos(b)]])
    result = np.matmul(rotMatrix, v)
    return result

rot3DVector(np.array([1, 1, 1]), 10, 10)
```
